```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
x = range(1000)
y = []
for i in range(1000):
    y.append(np.random.rand())
```

```python
plt.rcParams["font.family"] = "Helvetica"            #
plt.rcParams['mathtext.fontset'] = 'stix'            # math font

plt.rcParams["xtick.direction"] = "in"               #x
plt.rcParams["ytick.direction"] = "in"               #y
plt.rcParams["xtick.minor.visible"] = True           #x
plt.rcParams["ytick.minor.visible"] = True           #y
plt.rcParams["xtick.major.width"] = 1.5              #x
plt.rcParams["ytick.major.width"] = 1.5              #y
plt.rcParams["xtick.minor.width"] = 1.0              #x
plt.rcParams["ytick.minor.width"] = 1.0              #y
plt.rcParams["xtick.major.size"] = 10                #x
plt.rcParams["ytick.major.size"] = 10                #y
plt.rcParams["xtick.minor.size"] = 5                 #x
plt.rcParams["ytick.minor.size"] = 5                 #y
plt.rcParams["font.size"] = 14                       #
plt.rcParams["axes.linewidth"] = 1.5                 #

plt.rcParams['axes.grid'] = True                     # make grid

plt.rcParams["legend.fancybox"] = False              # 
plt.rcParams["legend.framealpha"] = 1                # 0
plt.rcParams["legend.edgecolor"] = 'black'           # edge
plt.rcParams["legend.handlelength"] = 1              # 
plt.rcParams["legend.labelspacing"] = 5.             # 
plt.rcParams["legend.handletextpad"] = 1.            # 
plt.rcParams["legend.markerscale"] = 2               # marker scale

#fig = plt.figure() # Figure
#ax = fig.add_subplot(1,1,1) # figAxes
plt.scatter(x, y, marker="o", alpha=0.5, color="b", label='test')

plt.title('test plot',fontsize=16)
plt.legend()

plt.xlabel('x',fontsize=14)
plt.ylabel('y',fontsize=14)

plt.xlim(0,1000)
plt.ylim(0,1)

plt.xticks(range(0,1000, 200), rotation=0, fontname='Arial',fontsize=10)
plt.yticks(np.arange(0, 1.1, 0.25),fontsize=10, rotation=0)

plt.savefig('test.png', dpi=500, bbox_inches="tight", pad_inches=0.05)
```
